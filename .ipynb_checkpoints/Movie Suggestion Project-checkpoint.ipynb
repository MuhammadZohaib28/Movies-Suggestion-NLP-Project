### Handle Genres

```python
movies.head(1)
```

```python
movies.iloc[0].genres
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def genresconversion(obj):
    emptyList = []
    for i in ast.literal_eval(obj):
        emptyList.append(i['name'])
    return emptyList    
```

```python
genresconversion('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(genresconversion)
```

```python
movies.head(2)
```

```python
movies.iloc[0].keywords
```

```python
movies['keywords'] = movies['keywords'].apply(genresconversion)
```

```python
movies.head(2)
```

```python
def collapsee(obj):
    emptyList = []
    for i in obj:
        emptyList.append(i.replace(" ", ""))
    return emptyList    
```

```python
movies['genres'] = movies['genres'].apply(collapsee)
movies['keywords'] = movies['keywords'].apply(collapsee)
```

```python
movies.head(2)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] =  movies['overview']+movies['genres']+movies['keywords']
```

```python
movies.head(1)
```

```python
movies.iloc[0].tags
```

```python
new  = movies.drop(columns=['overview', 'genres', 'keywords'])
```

```python
new.head(1)
```

```python
new['tags'] = new['tags'].apply(lambda x:" ".join(x))
```

```python
new.head(1)
```

```python
new.iloc[0].tags
```

```python
new['tags'] = new['tags'].apply(lambda x:x.lower())
```

```python
new.head(2)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
stem('In the 22nd century, a paraplegic Marine is dispatched to the moon Pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. Action Adventure Fantasy ScienceFiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d'
)
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
movies.head(1)
```

```python
new.head(1)
```

### Counter Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features= 5000, stop_words='english')
```

```python
vector =  cv.fit_transform(new['tags']).toarray()
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
len(similarity[0])
```

### Recommend

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key = lambda x:x[1])

    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
new[new['title'] == 'Avatar'].index[0]
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key = lambda x:x[1])
```

```python
recommend("Avatar")
```
